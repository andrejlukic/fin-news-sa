```python
'''
                    
                    Initialize sentiment analyzer
                    
'''
import nltk
import csv
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# load specialized stock market lexicon
stock_lex = pd.read_csv(r'.\lexicon_data\stock_lex.csv')
stock_lex['sentiment'] = (stock_lex['Aff_Score'] + stock_lex['Neg_Score'])/2
stock_lex = dict(zip(stock_lex.Item, stock_lex.sentiment))
stock_lex = {k:v for k,v in stock_lex.items() if len(k.split(' '))==1}
stock_lex_scaled = {}
for k, v in stock_lex.items():
    if v > 0:
        stock_lex_scaled[k] = v / max(stock_lex.values()) * 4
    else:
        stock_lex_scaled[k] = v / min(stock_lex.values()) * -4

# Update with Loughran and McDonald list of keywords
positive = []
with open(r'.\lexicon_data\lm_positive.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        positive.append(row[0].strip())
    
negative = []
with open(r'.\lexicon_data\lm_negative.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        entry = row[0].strip().split(" ")
        if len(entry) > 1:
            negative.extend(entry)
        else:
            negative.append(entry[0])

final_lex = {}
final_lex.update({word:2.0 for word in positive})
final_lex.update({word:-2.0 for word in negative})
final_lex.update(stock_lex_scaled)
final_lex.update(sia.lexicon)
sia = SentimentIntensityAnalyzer()
sia.lexicon = final_lex
```

```python
'''

            Load articles for INFY stock and calculate polarity for article title and for article body
            
'''

# stock market lexicon
i01 = pd.read_csv(r'.\scraped\test\bs.csv')
print(i01.size)
i02 = pd.read_csv(r'.\scraped\test\re.csv')
print(i02.size)
i03 = pd.read_csv(r'.\scraped\test\it.csv')
print(i03.size)
i01 = i01.append(i02).append(i03)

# some cleaning ...
i01 = i01.groupby('url', as_index=False).max()

i01['body'] = i01['body'].astype(str)
i01['sentiment_vader_LM_title'] = i01['title'].map(lambda a: sia.polarity_scores(a)['compound'])
i01['sentiment_vader_LM_body'] = i01['body'].map(lambda a: sia.polarity_scores(a)['compound'])
```

```python
'''
                                SIA 
'''
import datetime

# load historical stock data:
path_stock = r".\stocks\infy\\"
# read in all stock data files (each year is its own file) and concatenate the files.
all_files = glob.glob(os.path.join(path_stock, "*.csv"))
df_infy = pd.concat((pd.read_csv(f) for f in all_files))
df_infy['date'] = pd.to_datetime(df_infy['Date']).dt.date
df_infy = df_infy.sort_values(by='date', ascending=True)
df_infy.set_index(df_infy['date'], inplace=True)
df_infy.head()

start_date = datetime.date(2015,1,1)
end_date = datetime.date(2016,12,31)
mask = (df_infy['date'] > start_date) & (df_infy['date'] <= end_date)
df_infy_15_16 = df_infy[mask]
df_infy_15_16 = df_infy_15_16.set_index(df_infy_15_16['date'])
df_infy_15_16

trend1 = i01[['date','sentiment_vader_LM_title']]
trend1['date'] = pd.to_datetime(trend1['date']).dt.date
trend1['sentiment_vader_LM_title'] = pd.to_numeric(trend1['sentiment_vader_LM_title'])
trend1 = trend1.sort_values(by='date', ascending=True)

trend2 = i01[['date','sentiment_vader_LM_body']]
trend2['date'] = pd.to_datetime(trend2['date']).dt.date
trend2['sentiment_vader_LM_body'] = pd.to_numeric(trend2['sentiment_vader_LM_body'])
trend2 = trend2.sort_values(by='date', ascending=True)
```

```python
'''
    
            Sentiment: calculate average sentiment of the day
                       calculate simple signals based on sentiment
                        0 = no action
                       -1 = sell
                       +1 = buy
    
'''
import numpy as np

sig_title = trend2.groupby(trend2['date']).mean()
sig_title.sort_values(by='date').head()

sig_title = sig_title.join(df_infy[['Close Price','Low Price']], how='inner')
sa_column = 'sentiment_vader_LM_body'

sig_title["SA_Point"] = np.where(sig_title[sa_column] > 0.5, 1, sig_title[sa_column])
sig_title["SA_Point"] = np.where(sig_title[sa_column] < -0.1, -1, sig_title["SA_Point"])
sig_title["SA_Point"] = np.where(np.abs(sig_title["SA_Point"])!=1, 0, sig_title["SA_Point"])
sig_title['Signal'] = pd.Series()
sig_title['Profit'] = pd.Series()
last_action = None
for index, row in sig_title.iterrows():
    
    if(row["SA_Point"] == 0 or row["SA_Point"] == last_action):
        row['Signal'] = 0
    else:
        #print('{} last_action: {} new_action: {}'.format(index, last_action, row["SA_Point"]))
        row['Signal'] = last_action = row["SA_Point"]
        
        

sig_title_15_16 = sig_title[start_date:end_date]

#sig_title_15_16 = sig_title_15_16.join(df_infy_15_16['Close Price'], how='inner')
```

```python
'''
        
Prepare dataframe for backtesting for INFY stock

Price - Take closing price from where Signal is BUY
Profit column - a series of price differences (today - yesterday closing price)
End Date      - take date of the price of the day for which day before was Buy and Regime=1
''' 
sig_title_15_16 = sig_title_15_16[sig_title_15_16['Signal'] != 0]
sig_title_15_16['Profit'] = sig_title_15_16['Close Price'] - sig_title_15_16['Close Price'].shift(1)
sig_title_15_16.head()

sig_title_15_16 = sig_title_15_16[1:]
sig_title_15_16
```

```python
# ! TODO FIX: FIRST SIGNAL MUST BE + (BUY)

'''

    Backtesting general
    
'''

start_cash = 1000000
infy_backtest = pd.DataFrame({"Start Port. Value": [],
                         "End Port. Value": [],
                         "End Date": [],
                         "Shares": [],
                         "Share Price": [],
                         "Trade Value": [],
                         "Profit per Share": [],
                         "Total Profit": [],
                         "Stop-Loss Triggered": []})
port_value = 1  # Max proportion of portfolio bet on any trade
batch = 10      # Number of shares bought per batch
stoploss = .2    # % of trade loss that would trigger a stoploss
last_action = None
for index, row in sig_title_15_16.iterrows():
    
    signal = row['Signal']
    
    if(signal == -1 and last_action is not None):           
        stocks_available = int(last_action['Shares'])
        trade_val =  row["Close Price"] * stocks_available        
        profit = row['Profit'] * stocks_available  
        #print('sell {} {} {} ->{}'.format(index, stocks_available, trade_val, last_action["End Port. Value"]))
        # Add a row to the backtest data frame containing the results of the trade
        last_action = {
                    "Start Port. Value": last_action["End Port. Value"],
                    "End Port. Value": last_action["End Port. Value"] + trade_val,
                    # "End Date": row["End Date"],
                    "End Date": index,
                    "Shares": 0,
                    "Share Price": row["Close Price"],
                    "Trade Value": trade_val,
                    "Profit per Share": row['Profit'],
                    "Total Profit": profit,
                    "Stop-Loss Triggered": stop_trig
                }
        infy_backtest = infy_backtest.append(pd.DataFrame(last_action, index = [index]))        
    else:
        if(last_action is not None):
            cash = last_action['End Port. Value']
        else:
            cash = start_cash
        batches =  cash // np.ceil(batch * row["Close Price"]) # Maximum number of batches of stocks invested in
        trade_val = batches * batch * row["Close Price"] # How much money is put on the line with each trade
        shares = batches * batch
        #print('buy {} {} {} {} {}'.format(index, cash, trade_val, shares, cash - trade_val))
        if row["Low Price"] < (1 - stoploss) * row["Close Price"]:   # Account for the stop-loss
            share_profit = np.round((1 - stoploss) * row["Close Price"], 2)
            stop_trig = True
            print('stop triggered')
        else:            
            stop_trig = False
        
        share_profit = -row["Close Price"] 
        last_action = {
                    "Start Port. Value": cash,
                    "End Port. Value": cash - trade_val,
                    # "End Date": row["End Date"],
                    "End Date": index,
                    "Shares": shares,
                    "Share Price": row["Close Price"],
                    "Trade Value": trade_val,
                    "Profit per Share": share_profit,
                    "Total Profit": -trade_val,
                    "Stop-Loss Triggered": stop_trig
                }
        infy_backtest = infy_backtest.append(pd.DataFrame(last_action, index = [index]))        

infy_backtest
```
